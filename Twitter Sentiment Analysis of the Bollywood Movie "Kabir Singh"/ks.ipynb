### Now we will check the sentiments of the people through tweet text. 
#### *We will take the help of dictionary to check the polarity of words used in the tweet.

```python
dictionary = sqlContext.sql("SELECT * FROM dictionary")

dictionary.show(5)
```

### Different views are created to calculate the polarity in a series of steps

```python
%sql
create or replace view  v1 as
select author, texts
from kabir
lateral view explode(sentences(lower(text))) dummy as texts;

select * from v1 limit 5;
```

```python
%sql
create or replace view  v2 as
select author, word
from v1
lateral view explode(texts) dummy as word;

select * from v2 limit 5;
```

```python
%sql
create or replace view v3 as 
select author, v2.word,
case d.sentiment
     when 'negative' then -1
     when 'positive' then 1
     else 0 
end as polarity
from v2 
left outer join dictionary d on v2.word = d.word;

select * from v3 limit 5;
```

```python
%sql
create or replace view v4 as
select author, 
case
when sum( polarity ) > 0 then 'positive'
when sum( polarity ) < 0 then 'negative'
else 'neutral' end as sentiment
from v3 group by author;

select * from v4 limit 5;
```

## Pie Chart for the sentiment

```python
%sql
select sentiment, count(author) from v4 group by sentiment
```
